```python
# Dependencies
import requests
from config import api_key
```

```python
# Build query URL and request your results in Celsius

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=metric&APPID=" + api_key
print(url)

# Get weather data

# Create an endpoint URL for a city
city_url = url + "&q=" + "Bujumbura"
city_weather = requests.get(city_url)
city_weather = city_weather.json()
city_weather
```

```python
# Get temperature from JSON response

city_temp = city_weather["main"]["temp"]
```

```python
# Report temperature

print("The temperature (in degrees Celsius) in Bujumbura is " + str(city_temp) + ". The temperature (in degrees Fahrenheit) in Bujumbura is " + str(float(city_temp)* 9/5 + 32))
```
